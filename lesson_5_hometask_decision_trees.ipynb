```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import graphviz
import os
import xgboost

from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.datasets import load_wine

os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz/bin'
```

```python
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target
X = X.drop('od280/od315_of_diluted_wines', axis=1)

features = X.columns.tolist()
labels = wine.target_names
print ('labels:',labels)
print ('features:', features)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

Decision tree

```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
graph_viz = tree.export_graphviz(clf, out_file=None,feature_names=features, class_names=labels, filled = True)
graph = graphviz.Source(graph_viz)
graph.view(cleanup =True) # cleanup (bool) – Delete the source file after rendering.
```

Random Forrest

```python
from sklearn.ensemble import RandomForestClassifier
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=42)
clf.fit(X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```

Gradient Boosting Decision Trees

```python
from sklearn.ensemble import GradientBoostingClassifier
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = GradientBoostingClassifier(n_estimators=52, max_depth=2, learning_rate=0.01, random_state=42)
clf.fit(X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```

XGBoost

```python
from xgboost import XGBClassifier

# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = XGBClassifier(n_estimators=45, max_depth=2, learning_rate=0.03, reg_lambda=1, random_state=42, eval_metric='mlogloss')
clf.fit(X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```

Regression

```python
from sklearn.datasets import fetch_openml
data_reg = fetch_openml(name='house_prices', version=1, as_frame=True)
df_house = data_reg.frame
df_house
```

```python
from sklearn.ensemble import RandomForestRegressor
X_house = data_reg.data.select_dtypes(include=[np.number]).fillna(0)
y_house = data_reg.target
X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(X_house, y_house, random_state=42)
reg_forest = RandomForestRegressor(n_estimators=50, max_depth=9, random_state=42)
reg_forest.fit(X_train_h, y_train_h)
print("House Prices - Train R2 score: {:.3%}".format(reg_forest.score(X_train_h, y_train_h)))
print("House Prices - Test R2 score: {:.3%}".format(reg_forest.score(X_test_h, y_test_h)))
```
